```python
# Create stopword list:
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)
```

```python
# Note that 'text' is the name of the column in the DataFrame that contains the text data
# Generate wordcloud data
wordcloud = WordCloud(stopwords=stopwords, max_words=25, 
                      background_color="white").generate(' '.join(email_spam['text']))
```

## 01_03 Displaying the Wordcloud

```python
import matplotlib.pyplot as mpLib
mpLib.imshow(wordcloud)
mpLib.axis("off")
mpLib.show()
```

## 01_04 Enhancing Wordcloud

```python
# Assuming 'email_spam' is a DataFrame and the text is in a column named 'text_column'
# Convert the DataFrame column to a single string
email_text = ' '.join(email_spam['text'].astype(str))

# Add more words to ignore 
stopwords.update(["Customer Service", "information", "may", "Team"])

# Redo stop words. Limit number of words
wordcloud = WordCloud(stopwords=stopwords, max_words=10, 
                      background_color="azure").generate(email_text)

mpLib.imshow(wordcloud)
mpLib.axis("off")
mpLib.show()
```
